```
df_quant_sold = df[['Category','Quantity']].groupby('Category').sum()
df_quant_sold.reset_index(inplace = True)
#order descending
df_quant_sold.sort_values(by = ['Quantity'], ascending = False, inplace = True)
```

```
df_quant_sold.head()
```

```
df_segment_sold = df[['Segment','Quantity']].groupby('Segment').sum()
df_segment_sold.reset_index(inplace = True)
#order descending
df_segment_sold.sort_values(by = ['Quantity'], ascending = False, inplace = True)
```

```
df_segment_sold.head()
```

```
df_segment_profit = df[['Segment','Profit']].groupby('Segment').sum()
df_segment_profit.reset_index(inplace = True)
#order descending
df_segment_profit.sort_values(by = ['Profit'], ascending = False, inplace = True)
```

```
df_segment_profit.head()
```

```
df_state_sold = df[['State','Profit']].groupby('State').sum()
df_state_sold.reset_index(inplace = True)
#order descending
df_state_sold.sort_values(by = ['Profit'], ascending = False, inplace = True)
```

```
df_state_sold.head()
```

```
df_region_sold = df[['Region','Profit']].groupby('Region').sum()
df_region_sold.reset_index(inplace = True)
#order descending
df_region_sold.sort_values(by = ['Profit'], ascending = False, inplace = True)
```
